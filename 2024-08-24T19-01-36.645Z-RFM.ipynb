```python
#recenccy

recency=maxdate.drop('InvoiceDate',axis=1)
recency
```

```python
freq=df.drop_duplicates(subset="InvoiceNo")
freq
```

```python
freq = freq.groupby(["CustomerID"]).count()[["InvoiceNo"]]
freq
```

```python
#monetary
df["total"]=df["Quantity"]*df["UnitPrice"]
```

```python
df
```

```python
money=df.groupby(["CustomerID"])[["total"]].sum()
money
```

```python
df[df['CustomerID']==12346.0]
```

```python
RFM=pd.concat([recency,freq,money], axis=1)
RFM
```

```python
RFM.columns = ['Recency', 'Frequency', 'Monetary']
RFM
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaled=scaler.fit_transform(RFM)
```

```python
scaled
```

```python
scaled.shape
```

```python
from sklearn.cluster import KMeans
```

### To find out the value of K

```python
k_range = range(1,10)
sse = []
for k in k_range:
    km = KMeans(n_clusters=k) 
    km.fit(scaled)
    sse.append(km.inertia_)
```

```python
sse
```

```python
plt.plot(k_range, sse, 'r-*')
plt.xlabel ='k'
plt.ylabel = 'Sum of Squared Error'
```

### Object Creation and data fitting

```python
kmeans=KMeans(n_clusters=3)
kmeans.fit(scaled)
```

```python
kmeans.labels_
```

```python
kmeans.labels_.shape
```

```python
RFM["Clusters"]=(kmeans.labels_+1)
```

```python
RFM
```

```python
RFM[RFM.Clusters==3]
```

```python
final=RFM.groupby("Clusters").mean()[["Recency","Frequency","Monetary"]]
final
```

```python
def func(row):
    if row["Clusters"]==3:
        return 'Dimond'
    elif row["Clusters"]==2:
        return 'Gold'
    else:
        return 'Silver'
```

```python
RFM['group']=RFM.apply(func, axis=1)
```

```python
RFM
```

```python
result=RFM.group.value_counts()
```
